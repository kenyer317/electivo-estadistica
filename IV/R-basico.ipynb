## Medidas asociadas a datos cuantitativos


### Medidas de tendencia central
Indican dónde se sitúa el "centro" o el valor típico de los datos. Son el primer paso para entender la posición de la distribución.

1. **Media** $(\bar{x})$: El promedio aritmético. Es sensible a valores extremos (outliers).
$$\bar{x}= \frac{\text{ Suma de todos los datos}}{\text{Número de datos}}=\frac{x_1+x_2+\cdots+x_n}{n}$$


2. **Mediana**: El valor central que divide los datos al 50%. Es muy robusta frente a valores atípicos. Para su cálculo los datos deben estar ordenados y su formula es: 
$$ \text{Mediana} = \begin{cases}
\frac{x_{(\frac{n}{2})} + x_{(\frac{n}{2} + 1)}}{2} & \text{si } n \text{ es par} \\
x_{(\frac{n+1}{2})} & \text{si } n \text{ es impar}
\end{cases} $$


### Medidas de Dispersión (Variabilidad)
Indican qué tan alejados están los datos respecto al centro. Dos grupos pueden tener la misma media pero ser muy distintos si uno está más "disperso".


1. **Rango**: Diferencia entre el valor máximo y el mínimo.
$$\text{Rango} = \text{max} -\text{min}$$

2. **Varianza** : Promedio de los cuadrados de las desviaciones respecto a la media.
$$varianza = \frac{(x_1-\bar{x})^2+(x_2-\bar{x})^2+\cdots (x_n-\bar{x})^2}{n}$$

3. **Desviación Estándar**: La raíz cuadrada de la varianza. Expresa la dispersión en las mismas unidades que los datos originales (ej. puntos de calificación).
$$\text{desviacion estándar}=\sqrt{varianza}$$
 
4. **Desviación Media**: Es una medida de dispersión que nos indica, en promedio, qué tan alejados están los datos de su media aritmética. A diferencia de la varianza (que eleva las diferencias al cuadrado), la desviación media utiliza el valor absoluto para evitar que las diferencias positivas y negativas se cancelen entre sí.
$$\text{Desviacion Media}= \frac{|x_1-\bar{x}|+|x_2-\bar{x}|+\cdots +|x_n-\bar{x}|}{n}$$


5. **Coeficiente de Variación (CV)**: Útil para comparar la variabilidad entre grupos con distintas unidades o medias muy diferentes. Se puede expresar como porcentaje de la forma:
**Interpretación:**

$$CV = \frac{\text{desviación estándar}}{|\bar{x}|}\cdot 100\%$$
- Un coeficiente de variación bajo (cerca de 0) indica que los datos son relativamente homogéneos y tienen poca variabilidad en relación con la media.
-  Un coeficiente de variación alto sugiere que los datos son más dispersos y presentan gran variabilidad en comparación con la media.

### Medidas de Posición
Dividen la serie de datos en intervalos iguales. Son esenciales para entender la distribución de los estudiantes en percentiles.

1. Cuartiles $(Q_1, Q_2 ,Q_3)$: Dividen los datos en cuatro partes (25% cada una).

2. Quintiles y Deciles: Dividen en 5 y 10 partes respectivamente.

3. Percentiles $(P_k)$: Dividen los datos en 100 partes. Útiles para rankings (ej. "El estudiante está en el percentil 90").

## Ejemplo: Tiempo de uso del celular por día (en horas). (30 estudiantes)

``` 1.5, 2.0, 2.5, 3.0, 3.5,
4.0, 2.5, 3.0, 4.5, 5.0,
3.5, 2.0, 4.0, 3.0, 2.5,
5.5, 6.0, 4.5, 3.5, 2.5,
1.0, 2.0, 3.5, 4.0, 5.0,
3.0, 2.5, 4.5, 3.5, 2.0 
```

Se pide lo siguiente:
1. Guarda los datos en una variable llamada tiempo
2. Halla el maximo y el mínimo valor de la variable.
3. Halla el rango.
4. Halla el número de datos
5. Halla la media
6. Halla la mediana
7. Halla la varianza
8. Halla la desviación estándar
9. Halla la desviación media.
10. Halla los cuartiles $Q_1, Q_2, Q_3$.



### Solución

````r
# 1. Guarda los datos en una variable llamada tiempo
tiempo = c(
  1.5,
  2.0,
  2.5,
  3.0,
  3.5,
  4.0,
  2.5,
  3.0,
  4.5,
  5.0,
  3.5,
  2.0,
  4.0,
  3.0,
  2.5,
  5.5,
  6.0,
  4.5,
  3.5,
  2.5,
  1.0,
  2.0,
  3.5,
  4.0,
  5.0,
  3.0,
  2.5,
  4.5,
  3.5,
  2.0
)
tiempo
````

````r
# 2. Halla el maximo y el mínimo valor de la variable.
maximo = max(tiempo)
print(maximo)
minimo = min(tiempo)
print(minimo)
````

````r
# 3. Halla el rango.
rango = maximo - minimo
rango
````

````r
# 4. Halla el número de datos
numero_datos = length(tiempo)
numero_datos
````

````r
# 5. Halla la media
media = sum(tiempo) / numero_datos
media
````

````r
# 6. Halla la mediana
# Ordenamos los datos
tiempo = sort(tiempo)
# como n = 30 es par, la mediana es el promedio de los datos centrales de las posiciones 15 y 16
mediana = (tiempo[15] + tiempo[16]) / 2
mediana
````

````r
# 7. Halla la varianza
varianza = sum((tiempo - media)^2) / numero_datos
varianza
````

````r
# 8. Halla la desviación estándar
desviacion_estandar = sqrt(varianza)
desviacion_estandar
````

````r
# 9. Halla la desviación media.
desviacion_media = sum(abs(tiempo - media)) / numero_datos
desviacion_media
````

````r
# 10. Halla los cuartiles $Q_1, Q_2, Q_3$.
q1 = quantile(tiempo, 0.25)
print(q1)
q2 = quantile(tiempo, 0.50)
print(q2)
q3 = quantile(tiempo, 0.75)
print(q2)
````

# Covarianza

La covarianza es una medida estadística que indica cómo varían conjuntamente dos variables en una muestra. Sirve para detectar si existe una relación lineal entre ellas.

- Si es positiva → cuando una variable aumenta, la otra también tiende a aumentar.
- Si es negativa → cuando una variable aumenta, la otra tiende a disminuir.
- Si es cercana a 0 → no hay relación lineal clara.

$$\text{cov}(X,Y) = \frac{1}{n} \sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})$$





# Coeficiente de correlación de Pearson
$$r = \frac{\sum_{i=1}^n (x_i - \bar{x})(y_i - \bar{y})}{\sqrt{\sum_{i=1}^n (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^n (y_i - \bar{y})^2}}$$
Forma compacta
$$r = \frac{cov(X, Y)}{\sqrt{var(X)}\sqrt{var(Y)}}$$

## Interpretación 

- El valor de r nos dice:

- r=1 → correlación positiva perfecta
- r=−1 → correlación negativa perfecta
- r=0 → no hay relación lineal

## Ejemplos:

- Horas de estudio vs notas → correlación positiva
- Precio vs demanda → correlación negativa
- Número de zapatos vs inteligencia → ≈ 0 (sin relación)

````r
## Ejemplo en R
x <- c(2, 4, 6, 8, 10)
y <- c(1, 3, 5, 7, 9)
````

````r
# Covarianza
cov(x, y)
````

````r
# Coeficiente de correlación 
cor(x, y)
````

##  Importante
- cov() → depende de unidades
- cor() → siempre entre -1 y 1

## Recta de regresion
La recta de regresión lineal es una herramienta que permite modelar la relación entre dos variables y hacer predicciones. Básicamente, ajusta la “mejor” recta posible a un conjunto de datos. La ecuación de la recta de regresión lineal es:
$$y= bx+a$$
$$b= \frac{cov(X,Y)}{var(x)}$$
$$a = \bar{y}-b\bar{x}$$

### Interpretación
- b>0 → relación positiva
- b<0 → relación negativa
- b=0 → no hay relación lineal

````r
## Ejemplo en R


x <- c(2, 4, 6, 8, 10)
y <- c(1, 3, 5, 7, 9)

modelo <- lm(y ~ x)
modelo

plot(x, y)
abline(modelo, col="red")
````
