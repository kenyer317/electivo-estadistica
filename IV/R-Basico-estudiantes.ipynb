## Medidas asociadas a datos cuantitativos


### Medidas de tendencia central
Indican dónde se sitúa el "centro" o el valor típico de los datos. Son el primer paso para entender la posición de la distribución.

1. **Media** $(\bar{x})$: El promedio aritmético. Es sensible a valores extremos (outliers).
$$\bar{x}= \frac{\text{ Suma de todos los datos}}{\text{Número de datos}}=\frac{x_1+x_2+\cdots+x_n}{n}$$


2. **Mediana**: El valor central que divide los datos al 50%. Es muy robusta frente a valores atípicos. Para su cálculo los datos deben estar ordenados y su formula es: 
$$ \text{Mediana} = \begin{cases}
\frac{x_{(\frac{n}{2})} + x_{(\frac{n}{2} + 1)}}{2} & \text{si } n \text{ es par} \\
x_{(\frac{n+1}{2})} & \text{si } n \text{ es impar}
\end{cases} $$


### Medidas de Dispersión (Variabilidad)
Indican qué tan alejados están los datos respecto al centro. Dos grupos pueden tener la misma media pero ser muy distintos si uno está más "disperso".


1. **Rango**: Diferencia entre el valor máximo y el mínimo.
$$\text{Rango} = \text{max} -\text{min}$$

2. **Varianza** : Promedio de los cuadrados de las desviaciones respecto a la media.
$$varianza = \frac{(x_1-\bar{x})^2+(x_2-\bar{x})^2+\cdots (x_n-\bar{x})^2}{n}$$

3. **Desviación Estándar**: La raíz cuadrada de la varianza. Expresa la dispersión en las mismas unidades que los datos originales (ej. puntos de calificación).
$$\text{desviacion estándar}=\sqrt{varianza}$$
 
4. **Desviación Media**: Es una medida de dispersión que nos indica, en promedio, qué tan alejados están los datos de su media aritmética. A diferencia de la varianza (que eleva las diferencias al cuadrado), la desviación media utiliza el valor absoluto para evitar que las diferencias positivas y negativas se cancelen entre sí.
$$\text{Desviacion Media}= \frac{|x_1-\bar{x}|+|x_2-\bar{x}|+\cdots +|x_n-\bar{x}|}{n}$$


5. **Coeficiente de Variación (CV)**: Útil para comparar la variabilidad entre grupos con distintas unidades o medias muy diferentes. Se puede expresar como porcentaje de la forma:
**Interpretación:**

$$CV = \frac{\text{desviación estándar}}{|\bar{x}|}\cdot 100\%$$
- Un coeficiente de variación bajo (cerca de 0) indica que los datos son relativamente homogéneos y tienen poca variabilidad en relación con la media.
-  Un coeficiente de variación alto sugiere que los datos son más dispersos y presentan gran variabilidad en comparación con la media.

### Medidas de Posición
Dividen la serie de datos en intervalos iguales. Son esenciales para entender la distribución de los estudiantes en percentiles.

1. Cuartiles $(Q_1, Q_2 ,Q_3)$: Dividen los datos en cuatro partes (25% cada una).

2. Quintiles y Deciles: Dividen en 5 y 10 partes respectivamente.

3. Percentiles $(P_k)$: Dividen los datos en 100 partes. Útiles para rankings (ej. "El estudiante está en el percentil 90").
