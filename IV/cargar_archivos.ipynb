## Datos Desde Archivo CSV (.csv)

```r
datos = read.csv("ventas_electronica.csv")
datos
```

## Desde Archivo Excel (.xlsx)
### 1. Instalar la libreria  Readxl (sólo una vez): comando `install.packages("readxl")`

```r
install.packages("readxl")
```

### 2. Cargar librería y leer archivo

```r
library(readxl)
datos2= read_excel("ventas_excel.xlsx")
datos2
```
