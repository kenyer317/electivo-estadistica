```r
# Fijar semilla para reproducibilidad
set.seed(42)

# Número de registros
n <- 1000

# Listas de valores posibles
productos <- c("Smartphone", "Laptop", "Monitor", "Tablet", 
               "Audífonos", "Cámara Web", "Teclado Mecánico", "Reloj Inteligente")

categorias <- c(
  "Smartphone" = "Móvil",
  "Tablet" = "Móvil",
  "Laptop" = "Computación",
  "Monitor" = "Computación",
  "Audífonos" = "Audio",
  "Cámara Web" = "Accesorios",
  "Teclado Mecánico" = "Periféricos",
  "Reloj Inteligente" = "Wearables"
)

canales <- c("Online", "Tienda Física", "App")
metodos_pago <- c("Efectivo", "Tarjeta de Crédito", "Transferencia", "Mercado Pago")

# Fecha inicial
fecha_inicio <- as.Date("2023-01-01")

# Crear data frame vacío
df <- data.frame()

# Generación de datos
for (i in 1:n) {
  
  producto <- sample(productos, 1)
  categoria <- categorias[producto]
  cantidad <- sample(1:5, 1)
  
  precio <- round(runif(1, 50, 1300), 2)
  descuento <- sample(c(0, 0.05, 0.1, 0.15), 1)
  
  total <- round(cantidad * precio * (1 - descuento), 2)
  
  fila <- data.frame(
    ID_Venta = 1001 + (i - 1),
    Fecha = fecha_inicio + (2 * (i - 1)),
    Producto = producto,
    Categoria = categoria,
    Cantidad = cantidad,
    Precio_Unitario = precio,
    Descuento_Aplicado = descuento,
    Total_Venta = total,
    Canal_Venta = sample(canales, 1),
    Metodo_Pago = sample(metodos_pago, 1)
  )
  
  df <- rbind(df, fila)
}

# Guardar como CSV
write.csv(df, "ventas_1000_registros.csv", row.names = FALSE)

# Ver primeras filas
head(df)
```
